# clustering : univariate, bivariate and multivariate

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
clustering1 = KMeans(n_clusters=3)
```

```python
clustering1.fit(df[['Annual Income (k$)']])
```

```python
clustering1.labels_
```

```python
df['income cluster'] = clustering1.labels_
df.head()
```

```python
df['income cluster'].value_counts()
```

```python
clustering1.inertia_
```

```python
df['Annual Income (k$)'] = df['Annual Income (k$)'].astype(int)
```

```python
from sklearn.cluster import KMeans

inertia_score = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df[['Annual Income (k$)']])
    inertia_score.append(kmeans.inertia_)
```

```python
inertia_score
```

```python
plt.plot(range(1, 11), inertia_score)
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia Score')
plt.title('Inertia Score vs. Number of Clusters')
plt.show()
```

```python
#bivariate clustering 
```

```python
df.columns
```

```python
clustering2 = KMeans(n_clusters=5)
clustering2.fit(df[['Annual Income (k$)','Spending Score (1-100)']])
clustering2.labels_
df['income and spending cluster'] = clustering2.labels_
```

```python
df.head()
```

```python
from sklearn.cluster import KMeans

inertia_score2 = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df[['Annual Income (k$)','Spending Score (1-100)']])
    inertia_score2.append(kmeans.inertia_)
```

```python
inertia_score2
```

```python
plt.plot(range(1, 11), inertia_score2)
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia Score 2')
plt.title('Inertia Score vs. Number of Clusters')
plt.show()
```

```python
df.columns
```

```python
centers=pd.DataFrame(clustering2.cluster_centers_)
centers.columns =('x','y')
centers
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x=centers['x'], y=centers['y'], marker='*', c = 'black')
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)',hue='income and spending cluster', palette='tab10')
```
